```python
import os
import subprocess
from pathlib import Path

def parse_all_pdfs():
    """
    è‡ªå‹•å°‹æ‰¾ä¸¦ä½¿ç”¨ mineru è§£ææ‰€æœ‰ PDF æª”æ¡ˆã€‚
    """
    # --- è¨­å®š ---
    input_dir = Path("c:/Users/User/Desktop/GraphRAG/policy_dataset/AML_SPM")
    output_dir = Path("c:/Users/User/Desktop/GraphRAG/out_md_mineru")

    # ç¢ºä¿è¼¸å‡ºç›®éŒ„å­˜åœ¨
    output_dir.mkdir(exist_ok=True)

    print("ğŸ” é–‹å§‹è§£æ PDF æª”æ¡ˆ...")
    print(f"   ä¾†æº: {input_dir}")
    print(f"   è¼¸å‡º: {output_dir}")
    print("-" * 50)

    # éè¿´å°‹æ‰¾æ‰€æœ‰ .pdf æª”æ¡ˆ
    pdf_files = list(input_dir.glob("**/*.pdf"))
    
    if not pdf_files:
        print("âš ï¸ åœ¨ä¾†æºç›®éŒ„ä¸­æ‰¾ä¸åˆ°ä»»ä½• PDF æª”æ¡ˆã€‚")
        return

    for i, pdf_path in enumerate(pdf_files):
        print(f"ğŸ“„ ({i+1}/{len(pdf_files)}) æ­£åœ¨è™•ç†: {pdf_path.name}")
        
        # å»ºç«‹ä¸¦åŸ·è¡Œ mineru å‘½ä»¤
        command = [
            "mineru",
            "-p", str(pdf_path),
            "-o", str(output_dir)
        ]
        
        try:
            result = subprocess.run(command, check=True, capture_output=True, text=True, encoding='utf-8')
            print(result.stdout)
            print("âœ… å®Œæˆ.")
        except FileNotFoundError:
            print("âŒ éŒ¯èª¤: 'mineru' å‘½ä»¤æœªæ‰¾åˆ°ã€‚è«‹ç¢ºèªå·²å®‰è£ä¸¦åœ¨ç³»çµ± PATH ä¸­ã€‚")
            return
        except subprocess.CalledProcessError as e:
            print(f"âŒ è™•ç†å¤±æ•—: {pdf_path.name}")
            print(f"   éŒ¯èª¤è¨Šæ¯: {e.stderr}")
        
        print("")

    print("ğŸ‰ æ‰€æœ‰ PDF è§£æå®Œæˆï¼")

parse_all_pdfs()
```
